```python
merged = songs_df.merge(raaa, on= 'song_id', how = 'left')
merged.head(10)
```

```python
raaa.dtypes
```

```python
actual_data = merged[merged['avg'].notna()]
```

```python
actual_data.head(10)
```

```python
rat_counts = actual_data[actual_data['avg'].notnull()]['avg'].astype('int')
rat_counts = rat_counts*100
rat_averages = actual_data[actual_data['avg'].notnull()]['avg'].astype('int')
C = rat_averages.mean()
C
```

```python
m = rat_counts.quantile(0.95)
m
```

```python
qualified = actual_data[(actual_data['avg']*100 >= m) & (actual_data['avg'].notnull()) ][['title', 'year', 'release', 'avg', 'artist_name']]
qualified['rat_counts'] = qualified['avg'].astype('int')
qualified['rat_counts'] = qualified['rat_counts'] * 100
qualified['rat_averages'] = qualified['avg'].astype('int')
qualified.shape
```

```python
qualified.head(10)
```

```python
def weighted_rating(x):
    v = x['rat_counts']
    R = x['rat_averages']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
qualified['wr'] = qualified.apply(weighted_rating, axis=1)
```

```python
qualified = qualified.sort_values('wr', ascending=False).head(250)
```

```python
qualified.head(15)
```

```python
def build_chart(artist, percentile=0.85):
    df = actual_data[actual_data['artist_name'] == artist]
    rat_counts = df[df['avg'].notnull()]['avg'].astype('int')
    rat_counts = rat_counts*100
    rat_averages = df[df['avg'].notnull()]['avg'].astype('int')
    C = rat_averages.mean()
    m = rat_counts.quantile(percentile)
    
    qualified = df[(df['avg']*100 >= m) & (df['avg'].notnull())][['title', 'year', 'release', 'avg', 'artist_name']]
    qualified['rat_counts'] = qualified['avg'].astype('int')
    qualified['rat_counts'] = qualified['rat_counts'] * 100
    qualified['rat_averages'] = qualified['avg'].astype('int')
    
    qualified['wr'] = qualified.apply(lambda x: (x['rat_counts']/(x['rat_counts']+m) * x['rat_averages']) + (m/(m+x['rat_counts']) * C), axis=1)
    qualified = qualified.sort_values('wr', ascending=False).head(250)
    
    return qualified
```
